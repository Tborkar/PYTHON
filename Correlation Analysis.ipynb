# Correlation Analysis visuals

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"T:\Data Field\PROJECT PORTFOLIO\Combine Software project\python\ev_data_statistics.xlsx")
pd.set_option('display.max.rows',104)
pd.set_option('display.max.columns',15)
df
```

## Price Variation based on Performance Metrics of EV car


```python
# features: feature1, feature2, feature3, feature4, feature5
# output: output_feature

feature1 = df['Accel']
feature2 = df['TopSpeed']
feature3 = df['Range']
feature4 = df['Efficiency']
feature5 = df['FastCharge']
output_feature = df['PriceRupees']

# Create subplots for each feature
fig, axs = plt.subplots(1, 5, figsize=(20, 4))

# Scatter plot for each feature against output_feature
for i, feature in enumerate([feature1, feature2, feature3, feature4, feature5]):
    axs[i].scatter(feature, output_feature, marker = "*", color = "red")
    axs[i].set_xlabel('Feature ' + str(i+1))
    axs[i].set_ylabel('Output Feature')

# Set labels for each subplot individually
axs[0].set_xlabel("Accel")
axs[1].set_xlabel("Top Speed")
axs[2].set_xlabel("Range")
axs[3].set_xlabel("Efficiency")
axs[4].set_xlabel("Fast Charge")

axs[0].set_ylabel("Price Rupees")
axs[1].set_ylabel("Price Rupees")
axs[2].set_ylabel("Price Rupees")
axs[3].set_ylabel("Price Rupees")
axs[4].set_ylabel("Price Rupees")

plt.tight_layout()
plt.show()
```

### Insight from above scatterplot is given below:
#### The analysis of the scatter plot reveals a discernible trend where a decrease in acceleration corresponds to an increase in price. Conversely, an increase in attributes such as top speed, range, efficiency, and fast charge demonstrates a proportional rise in price.

# Finding Correlation between feature using Pearson correlation coefficient

```python
import seaborn as sns

# df is DataFrame

# DataFrame contains non-numeric columns, so select only numeric columns for correlation
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
corr = df_numeric.corr()

# Plot correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of EV Data')
plt.show()
```
