# # Finding the outliers using the IQR, Upper Bound and Lower Bound

```python
# Calculate the IQR for each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Set the threshold for outlier detection
threshold = 1.5

# Find the indices of outliers
outlier_indices = (df < (Q1 - threshold * IQR)) | (df > (Q3 + threshold * IQR))

# Get the rows containing outliers
outliers = df[outlier_indices.any(axis=1)]

print("Outliers:")
print(outliers)
```

```python
import matplotlib.pyplot as plt


# Function to detect outliers using IQR
def detect_outliers_iqr(df):
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    outliers = []
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        col_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outliers.append(col_outliers)
    return pd.concat(outliers)


# Detect outliers
outliers_iqr = detect_outliers_iqr(df)

# Visualize outliers using box plots
plt.figure(figsize=(20, 15))
plt.subplot(1, 2, 1)
df.boxplot()
plt.title("Box Plot of Original Data")

plt.subplot(1, 2, 2)
outliers_iqr.boxplot()
plt.title("Box Plot of Outliers Detected by IQR")

plt.show()
```
