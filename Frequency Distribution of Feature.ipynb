## Pareto Analysis of Brand of EV Car 

```python
# Perform value counts on the Brand column
brand_counts = df['Brand'].value_counts()

# Calculate cumulative percentage
brand_cumulative_percentage = brand_counts.cumsum() / brand_counts.sum() * 100

# Plot the Pareto chart
fig, ax1 = plt.subplots(figsize=(24, 6))  # Adjust the figure size

# Plot the bar plot
ax1.bar(brand_counts.index, brand_counts, color='y')
ax1.set_xlabel('Brand', fontsize=12)  # Increase font size of x-axis labels
ax1.set_ylabel('Frequency', color='black', fontsize=18)

# Create a secondary y-axis for cumulative percentage
ax2 = ax1.twinx()
ax2.plot(brand_counts.index, brand_cumulative_percentage, color='r', marker='o', linestyle='-')
ax2.set_ylabel('Cumulative Percentage (%)', color='r', fontsize=12)

# Add a horizontal line at 80% of the dataset
ax2.axhline(y=80, color='g', linestyle='--', label='80% of Models data')

# Calculate the position of the 80th percentile
percentile_80 = brand_cumulative_percentage[brand_cumulative_percentage >= 80].index[0]

# Add a vertical line at the 80th percentile
ax2.axvline(x=percentile_80, color='orange', linestyle='--', label='80th percentile')

# Show the plot
plt.title('80/20 Classification of Brand Data (Pareto Analysis)', fontsize=20)
plt.xticks(rotation=180,ha='right', fontsize=10)  # Rotate x-axis labels and increase font size
plt.tight_layout()
plt.legend(loc='center right')
plt.show()
```
